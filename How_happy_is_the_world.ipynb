The Sustainable Development Solutions Network (SDSN) collects data across the world relating to happiness.  They use this data to rank countries in order of happiness factor.

This is not an exact science but can give food for thought in terms of what factors might have the most impact on a nation's happiness levels.

Data is taken from the Gallup World Poll, so not collected directly by SDSN.  
Countries are grouped by region.  

### The factors included are:
Economy (measured in GDP per Capita)
Family (support systems)
Health (measured by Life Expectancy)
Freedom (sense of)
Trust (Government Corruption)
Generosity (charitable inclinations)
Dystopia Residual
*  Dystopic is the theoretical most unhappy country with the lowest levels in all six of the above factors  
*  The Residual measure is a calculated as the average of the six distances from lowest

Let's take a look at the data

### Open a data set

Open the data set, an Excel file with only one sheet (so sheet_name is not necessary) from here: https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true

Interrogate the data (head, tail, iloc) to get to know what it contains.


```
import pandas as pd
happiness = pd.read_excel("https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true")
happiness.head(10)
happiness.tail(10)
happiness.iloc[int(len(happiness)/2)-5:int(len(happiness)/2)+5]
```

---
### Sort the data in different ways

The data is currently sorted in order of rank.  To sort the data in the table, run the code below, which identifies the column on which to sort in the brackets.

Then, **try sorting on other columns** *Note: you must type the column heading in the quotes and exactly as it appears in the table (including capitalisation)*.  To sort on multiple columns, enter a list of column headings in the brackets (e.g. `.sort_values(['Region','Freedom'])`



```
sorted_table = happiness.sort_values(['Family'])
sorted_table1 = happiness.sort_values(["Happiness Rank"])
sorted_table2 = happiness.sort_values(["Economy (GDP per Capita)"])
sorted_table3 = happiness.sort_values(["Health (Life Expectancy)"])
sorted_table4 = happiness.sort_values(["Generosity"])
sorted_table5 = happiness.sort_values(["Trust (Government Corruption)"])
sorted_table6 = happiness.sort_values(["Freedom"])
sorted_table7 = happiness.sort_values(["Dystopia Residual"])
print(sorted_table)
print(sorted_table1)
print(sorted_table2)
print(sorted_table3)
print(sorted_table4)
print(sorted_table5)
print(sorted_table6)
print(sorted_table7)
```

### Summarising the data

Look at the happiness dataframe.  Create new dataframes from a range of rows, columns, statistical information, etc.

The two tables below show top 10 and bottom 10 countries by the level of freedom.

```
a = happiness[["Country", "Freedom"]].sort_values(["Freedom"], ascending=False)
print(a.head(10))
print(a.tail(10))
```

The two tables below show top 10 and bottom 10 countries by the level of trust in government.

```
b = happiness[["Country", "Trust (Government Corruption)"]].sort_values(["Trust (Government Corruption)"], ascending=False)
print(b.head(10))
print(b.tail(10))
```

The tables below display top 10 countries in the world by the level of generocity and the bottom 10.

```
c = happiness[["Country", "Generosity"]].sort_values("Generosity", ascending=False)
print(c.head(10))
print(c.tail(10))
```

The two tables printed below show the top 10 and bottom 10 countries by life expectancy.

```
d = happiness[["Country", "Health (Life Expectancy)"]].sort_values(["Health (Life Expectancy)"], ascending=False)
print(d.head(10))
print(d.tail(10))
```

The two tables below display the top ten and bottom ten countries by GDP per capita.

```
e = happiness[["Country", "Economy (GDP per Capita)"]].sort_values(["Economy (GDP per Capita)"], ascending=False)
print(e.head(10))
print(e.tail(10))
```

The two tables below display the top 10 and bottom 10 countries by the happiness score.

```
f = happiness[["Country", "Happiness Score"]].sort_values(["Happiness Score"], ascending=False)
print(f.head(10))
print(f.tail(10))
```

The two tables below display the top 10 countries with the lowest dystopia residual and the bottom 10 countries with the highest dystopia residual.

```
g = happiness[["Country", "Dystopia Residual"]].sort_values(["Dystopia Residual"])
print(g.head(10))
print(g.tail(10))
```

The two tables below show the top 10 and bottom 10 countries by the criterion of family.

```
h = happiness[["Country", "Family"]].sort_values(["Family"], ascending=False)
print(h.head(10))
print(h.tail(10))
```

The table below displays the highest, the lowest and the median score in each category.

```
i = happiness.agg(
    {
        "Happiness Score": ["max", "min", "median"],
        "Economy (GDP per Capita)": ["max", "min", "median"],
        "Family": ["max", "min", "median"],
        "Health (Life Expectancy)": ["max", "min", "median"],
        "Freedom": ["max", "min", "median"],
        "Trust (Government Corruption)": ["max", "min", "median"],
        "Generosity": ["max", "min", "median"],
        "Dystopia Residual": ["max", "min", "median"]


    }
)
print(i)
```

The table below shows the highest and lowest happiness score, GDP per capita, family score, life expectancy, freedom score, trust in government score, generosity score and dystopia residual score in each region.

```
j = happiness.groupby("Region").agg(
    {
        "Happiness Score": ["min", "max"],
        "Economy (GDP per Capita)": ["min", "max"],
        "Family": ["min", "max"],
        "Health (Life Expectancy)": ["min", "max"],
        "Freedom": ["min", "max"],
        "Trust (Government Corruption)": ["min", "max"],
        "Generosity": ["min", "max"],
        "Dystopia Residual": ["min", "max"]


    }
)
print(j)
```
