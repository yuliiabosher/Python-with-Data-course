# Data investigation project
---

Use what you have learnt so far about reading data files and displaying rows, columns, unique values, etc, to investigate and produce some printed information from this data set: https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv

```python
import pandas as pd
doc = pd.read_csv("https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv")
print(f'This document contains {doc.shape[0]} rows and {doc.shape[1]} columns')
print(f'This file contains {len(doc["code"].unique())} unique ONS codes')
print(f'The first 10 rows of the dataset are as follows \n\n{doc.head(10)}')
print(f'\n\nThe middle 10 rows of the dataset are as follows \n\n{doc.iloc[len(doc)//2-5:len(doc)//2+5]}')
print(f'\n\nThe last 10 rows of the dataset are as follows \n\n{doc.tail(10)}')
print("\nThe unique area names represented in this dataset are as follows \n")
for i in doc["area"].unique():
  print(i.capitalize())
print(f'\nThere are {len(doc["area"].unique())} distinct area names represented in this dataset')



```
